```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mhealth = pandas.read_table('mental_health_and_technology_usage_2024.csv')
```

```python
data = {
    'Screen_Time_Hours': np.random.uniform(1, 10, 100),  
    'Mental_Health_Score': np.random.uniform(1, 10, 100),  
    'Job_Satisfaction_Score': np.random.uniform(1, 10, 100)  
}
```

```python
health_frame = pd.DataFrame(data)

health_frame
```

```python
info_stats = health_frame.groupby('Mental_Health_Score').describe()
print(info_stats)
```

```python
plt.subplot(1, 2, 1)
sns.regplot(x='Screen_Time_Hours', y='Mental_Health_Score', data=health_frame, scatter_kws={'s': 20}, color='blue')
plt.title('Screen Time vs. Mental Health Score')
plt.xlabel('Screen Time (Hours)')
plt.ylabel('Mental Health Score')
```

```python
plt.subplot(1, 2, 2)
sns.regplot(x='Screen_Time_Hours', y='Job_Satisfaction_Score', data=health_frame, scatter_kws={'s': 20}, color='green')
plt.title('Screen Time vs. Job Satisfaction Score')
plt.xlabel('Screen Time (Hours)')
plt.ylabel('Job Satisfaction Score')
```
